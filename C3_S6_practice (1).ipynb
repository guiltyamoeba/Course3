```python
#task4
class Student:
    def __init__(self, enrollment_ID, student_name, course_name, age, marks):
        self.enrollment_ID = enrollment_ID
        self.student_name = student_name
        self.course_name = course_name
        self.age = age

        self.age_validation(age)
        self.marks = self.validate_and_convert_marks(marks)
        self.grade = self.calculate_grade(self.marks)

    def age_validation(self, age):
        if not isinstance(age, int):
            raise ValueError("Age must be an integer")
        if age < 3 or age > 60:
            raise ValueError("Age must be between 3 and 60")
        self.age = age

    def validate_and_convert_marks(self, marks):
        if not isinstance(marks, int):
            return "Invalid Marks"
        if 0 <= marks <= 44:
            return marks / 44 * 100  # Convert to percentage
        else:
            return "Invalid Marks"

    def calculate_grade(self, percentage):
        if percentage < 0:
            return "Invalid"
        elif percentage > 100:
            return "Impossible"
        elif percentage >= 80.01:
            return "Grade A"
        elif percentage >= 60.01:
            return "Grade B"
        elif percentage >= 40.01:
            return "Grade C"
        elif percentage >= 20.01:
            return "Grade D"
        else:
            return "Grade F"
```
